### Paradigm Identification Guide -- Summary

| Paradigm | When to Use |
|----------|-------------|
| **Brute Force** | Constraints are small (n <= a few hundred); "enumerate all cases" |
| **Divide & Conquer** | Problem splits into **independent** subproblems; no overlap |
| **Greedy** | "Best local choice = global optimum"; greedy choice property holds |
| **DP** | Optimal substructure + **overlapping** subproblems; recurrence relation exists |

**Decision flow:**
1. Read problem -> Check constraints (small? -> Brute Force)
2. Can it be split into independent parts? -> Divide & Conquer
3. Does best-at-each-step = overall best? -> Greedy
4. Overlapping subproblems + optimal substructure? -> DP

````python
# Solution: Merge Sort
def merge_sort_solution(arr):
    """Merge Sort solution.

    Time complexity: O(n log n)
    Space complexity: O(n)
    """
    merge_count = [0]  # Wrap in list so inner function can modify it

    def merge_sort(a):
        if len(a) <= 1:
            return a

        mid = len(a) // 2
        left = merge_sort(a[:mid])
        right = merge_sort(a[mid:])

        return merge(left, right)

    def merge(left, right):
        merge_count[0] += 1
        result = []
        i = j = 0

        while i < len(left) and j < len(right):
            if left[i] <= right[j]:  # <= for stability
                result.append(left[i])
                i += 1
            else:
                result.append(right[j])
                j += 1

        result.extend(left[i:])
        result.extend(right[j:])
        return result

    sorted_arr = merge_sort(arr)
    return sorted_arr, merge_count[0]


# Run solution
test_arr = [38, 27, 43, 3, 9, 82, 10]
sorted_arr, count = merge_sort_solution(test_arr[:])
print(f"Input:  {test_arr}")
print(f"Sorted: {sorted_arr}")
print(f"Merge count: {count}")
````

````python
# Solution: Divide and Conquer Maximum
def find_max_solution(arr):
    """Divide and conquer max-finding solution.

    Time complexity: O(n)
    Comparisons: n - 1 (optimal)
    T(n) = 2T(n/2) + 1 -> O(n)
    """
    comparisons = [0]

    def find_max(a, lo, hi):
        if lo == hi:
            return a[lo]

        mid = (lo + hi) // 2
        left_max = find_max(a, lo, mid)
        right_max = find_max(a, mid + 1, hi)

        comparisons[0] += 1
        return left_max if left_max >= right_max else right_max

    if not arr:
        return None, 0

    result = find_max(arr, 0, len(arr) - 1)
    return result, comparisons[0]


# Run solution
test_arr = [3, 7, 2, 9, 1, 8, 4, 6, 5]
max_val, comparisons = find_max_solution(test_arr)
print(f"Input: {test_arr}")
print(f"Maximum: {max_val}")
print(f"Comparisons: {comparisons} (theoretical minimum: {len(test_arr) - 1})")
````

````python
# Solution: Activity Selection
def activity_selection_solution(lectures):
    """Activity Selection solution.

    Greedy strategy: Sort by end time, pick non-overlapping.
    Time complexity: O(n log n)
    """
    # (end_time, start_time, original_index) for sorting
    indexed = [(end, start, i) for i, (start, end) in enumerate(lectures)]
    indexed.sort()

    selected = []
    last_end = -1

    for end, start, idx in indexed:
        if start >= last_end:
            selected.append(idx)
            last_end = end

    return len(selected), selected


# Run solution
lectures = [(1, 3), (2, 5), (3, 6), (5, 7), (6, 8), (8, 10), (9, 11)]
count, selected = activity_selection_solution(lectures)
print(f"Lectures: {lectures}")
print(f"Max lectures: {count}")
print(f"Selected: {[lectures[i] for i in selected]}")
````

````python
# Solution: Climbing Stairs with Cost
def climbing_stairs_solution(cost):
    """Climbing Stairs solution.

    dp[i] = cost[i] + min(dp[i-1], dp[i-2])

    Time complexity: O(n)
    Space complexity: O(1) (optimized)
    """
    n = len(cost)
    if n == 0:
        return 0
    if n == 1:
        return cost[0]

    # Space-optimized version
    prev2 = cost[0]
    prev1 = cost[1]

    for i in range(2, n):
        current = cost[i] + min(prev1, prev2)
        prev2, prev1 = prev1, current

    # Reach the top: jump from last stair or second-to-last stair
    return min(prev1, prev2)


# Run solution
cost1 = [10, 15, 20]
cost2 = [1, 100, 1, 1, 1, 100, 1, 1, 100, 1]
print(f"cost = {cost1} -> min cost: {climbing_stairs_solution(cost1)}")
print(f"cost = {cost2} -> min cost: {climbing_stairs_solution(cost2)}")
````
