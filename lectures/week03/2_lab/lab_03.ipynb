## A-1: Basic Sorting Algorithms (10 min)

All three basic sorts have **O(n²)** worst-case time complexity, but differ in how they work.
Each function takes a list and returns a **new** sorted list (does not modify the original).

### Selection Sort

**Idea**: Find the minimum element in the unsorted portion and swap it to the front.

```
[64, 34, 25, 12, 22, 11, 90]
 ^                   ^          Find min (11), swap with position 0
[11, 34, 25, 12, 22, 64, 90]
     ^       ^                  Find min (12), swap with position 1
[11, 12, 25, 34, 22, 64, 90]   ... and so on
```

````python
def selection_sort(arr):
    """Selection Sort: find minimum, swap to front. O(n^2)"""
    a = arr[:]
    n = len(a)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if a[j] < a[min_idx]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]
    return a


test = [64, 34, 25, 12, 22, 11, 90]
print(f"Original:       {test}")
print(f"Selection Sort: {selection_sort(test)}")
````

### Bubble Sort

**Idea**: Repeatedly swap adjacent elements if they are in the wrong order. After each pass, the largest unsorted element "bubbles up" to its correct position.

**Optimization**: If no swaps occur during a pass, the array is already sorted — we can stop early.

````python
def bubble_sort(arr):
    """Bubble Sort: repeatedly swap adjacent elements. O(n^2)"""
    a = arr[:]
    n = len(a)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
                swapped = True
        if not swapped:
            break
    return a


test = [64, 34, 25, 12, 22, 11, 90]
print(f"Original:    {test}")
print(f"Bubble Sort: {bubble_sort(test)}")
````

### Insertion Sort

**Idea**: Build the sorted array one element at a time. Take each element and insert it into the correct position among the already-sorted elements to its left.

**Note**: Insertion sort is efficient for nearly-sorted data and small arrays.

````python
def insertion_sort(arr):
    """Insertion Sort: insert each element into sorted prefix. O(n^2)"""
    a = arr[:]
    for i in range(1, len(a)):
        key = a[i]
        j = i - 1
        while j >= 0 and a[j] > key:
            a[j + 1] = a[j]
            j -= 1
        a[j + 1] = key
    return a


test = [64, 34, 25, 12, 22, 11, 90]
print(f"Original:       {test}")
print(f"Insertion Sort: {insertion_sort(test)}")
````

## A-2: Advanced Sorting Algorithms (15 min)

These algorithms achieve **O(n log n)** time complexity using divide-and-conquer.

### Merge Sort

**Idea**: Divide the array in half, recursively sort each half, then merge the two sorted halves.

```
[38, 27, 43, 3, 9, 82, 10]
      /                \
[38, 27, 43, 3]    [9, 82, 10]
    /       \         /      \
[38,27]  [43,3]    [9,82]  [10]
  / \     / \       / \
[38][27][43][3]   [9][82]
  \ /     \ /       \ /
[27,38]  [3,43]   [9,82]  [10]
    \       /         \      /
[3, 27, 38, 43]    [9, 10, 82]
      \                /
[3, 9, 10, 27, 38, 43, 82]
```

**Time**: O(n log n) always  |  **Space**: O(n) for the merge step

````python
def merge_sort(arr):
    """Merge Sort: divide, sort halves, merge. O(n log n)"""
    if len(arr) <= 1:
        return arr[:]
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return _merge(left, right)


def _merge(left, right):
    """Merge two sorted arrays into one sorted array."""
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result


test = [38, 27, 43, 3, 9, 82, 10]
print(f"Original:   {test}")
print(f"Merge Sort: {merge_sort(test)}")
````

### Quick Sort

**Idea**: Choose a pivot element, partition the array into elements less than, equal to, and greater than the pivot, then recursively sort the partitions.

**Time**: O(n log n) average, O(n²) worst case  |  **Space**: O(log n) for recursion

````python
def quick_sort(arr):
    """Quick Sort: partition around pivot, sort subarrays. O(n log n) avg."""
    if len(arr) <= 1:
        return arr[:]
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)


test = [38, 27, 43, 3, 9, 82, 10]
print(f"Original:   {test}")
print(f"Quick Sort: {quick_sort(test)}")
````

## A-3: Benchmark All Sorting Algorithms (10 min)

Now let's compare the performance of all five algorithms plus Python's built-in `sorted()`.

Note: Selection Sort and Bubble Sort are skipped for N > 5,000 because they are too slow.

````python
def measure(func, data, repeat=1):
    """Measure average execution time."""
    times = []
    for _ in range(repeat):
        start = time.perf_counter()
        func(data)
        end = time.perf_counter()
        times.append(end - start)
    return sum(times) / len(times)


sizes = [100, 500, 1000, 5000, 10000]
algorithms = {
    "Selection": selection_sort,
    "Bubble": bubble_sort,
    "Insertion": insertion_sort,
    "Merge": merge_sort,
    "Quick": quick_sort,
    "Python sorted()": sorted,
}
results = {name: [] for name in algorithms}

for n in sizes:
    data = [random.randint(1, 100000) for _ in range(n)]
    print(f"\nN = {n:,}")
    for name, func in algorithms.items():
        if n > 5000 and name in ("Selection", "Bubble"):
            results[name].append(None)
            print(f"  {name:>15}: skipped (too slow)")
            continue
        t = measure(func, data)
        results[name].append(t)
        print(f"  {name:>15}: {t:.6f} sec")
````

### Visualize the Benchmark Results

````python
if HAS_MATPLOTLIB:
    plt.figure(figsize=(10, 6))
    for name, times in results.items():
        valid = [(s, t) for s, t in zip(sizes, times) if t is not None]
        if valid:
            ss, ts = zip(*valid)
            plt.plot(ss, ts, marker='o', label=name)
    plt.xlabel("Input Size (N)")
    plt.ylabel("Time (seconds)")
    plt.title("Sorting Algorithm Benchmark")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("Install matplotlib to see the benchmark graph.")
````
